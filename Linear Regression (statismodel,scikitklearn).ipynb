## Using Stats model pkg 

```python
import statsmodels.api as sm
```

```python
data.head()
```

### Forward selection 

```python
Y = data['sales'] #depending variable
```

```python
X = data['TV'] #independing variable 
```

```python
#Add the constant 
X = sm.add_constant(X)
print(X)
```

```python
model1 = sm.OLS(Y,X).fit()

Y_pred = model1.predict(X)

print(model1.summary())
```

```python
#Distribution of Residuals 
plt.hist(model1.resid_pearson)
plt.ylabel('Count')
plt.xlabel('Normalized residuals')
```

```python
pd.DataFrame(model1.params)
```

```python
eval_metric = ['AIC','BIC','R-sqauared','Adj. R-squared','RMSE','VIF']
```

```python
model1_metrics = [model1.aic,model1.bic,model1.rsquared,model1.rsquared_adj,np.sqrt(model1.mse_resid),1/(1-model1.rsquared)]
```

```python
model1_metrics
```

# Using Scikit-learn

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
lm = LinearRegression()
lm.fit(X,Y)
```

```python
b0 = lm.intercept_
b1 = lm.coef_[1]
print(b0,b1)
```

```python
Y_predict = lm.predict(X)
```

```python
Y_predict
```

```python
print('R-Sqaured: ',metrics.r2_score(Y,Y_predict))
print('Root Mean Squared Error: ',np.sqrt(metrics.mean_squared_error(Y,Y_predict)))
```
